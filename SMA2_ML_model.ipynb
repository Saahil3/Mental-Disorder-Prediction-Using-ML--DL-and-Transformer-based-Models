```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.utils import resample
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from transformers import BertTokenizer, BertForSequenceClassification, Trainer, TrainingArguments
import torch


print("Initial Dataset Shape:", df.shape)
df.dropna(subset=['clean_text'], inplace=True)
print("After Dropping Missing Values:", df.shape)

encoder = LabelEncoder()
df['disorder_encoded'] = encoder.fit_transform(df['disorder'])

balanced_df = []
for label in df['disorder_encoded'].unique():
    samples = df[df['disorder_encoded'] == label]
    balanced_samples = resample(samples, replace=True, n_samples=4000, random_state=42)
    balanced_df.append(balanced_samples)

df_balanced = pd.concat(balanced_df)
print("After Balancing Dataset:", df_balanced.shape)

X_train, X_test, y_train, y_test = train_test_split(df_balanced['clean_text'], df_balanced['disorder_encoded'],
                                                    test_size=0.2, random_state=42)
print("Training Set Size:", X_train.shape)
print("Test Set Size:", X_test.shape)

vectorizer = TfidfVectorizer(max_features=5000)
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)

rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train_tfidf, y_train)
rf_predictions = rf_model.predict(X_test_tfidf)

nb_model = MultinomialNB()
nb_model.fit(X_train_tfidf, y_train)
nb_predictions = nb_model.predict(X_test_tfidf)

def evaluate_model(predictions, model_name):
    print(f"\nPerformance of {model_name}")
    print(classification_report(y_test, predictions, target_names=encoder.classes_))
    sns.heatmap(confusion_matrix(y_test, predictions), annot=True, fmt="d", cmap="Blues")
    plt.title(f"Confusion Matrix - {model_name}")
    plt.show()

evaluate_model(rf_predictions, "Random Forest")
evaluate_model(nb_predictions, "Multinomial Naive Bayes")
```

```python
import joblib
import pandas as pd
import matplotlib.pyplot as plt
from ipywidgets import widgets
from IPython.display import display, clear_output


rf_model = joblib.load("random_forest_model.pkl")
vectorizer = joblib.load("tfidf_vectorizer.pkl")
label_encoder = joblib.load("label_encoder.pkl")

def create_rf_prediction_tool(model, vectorizer, encoder):
    """
    Create an interactive prediction tool using a Random Forest model with TF-IDF features.
    """
    print("Creating interactive prediction tool...")

    
    text_input = widgets.Textarea(
        value='',
        placeholder='Enter text to classify...',
        description='Input:',
        disabled=False,
        layout={'width': '100%', 'height': '100px'}
    )

    analyze_button = widgets.Button(
        description='Analyze',
        button_style='success',
        tooltip='Click to analyze the text'
    )

    output = widgets.Output()

    def on_button_click(b):
        with output:
            clear_output()
            input_text = text_input.value.strip()
            if not input_text:
                print("Please enter some text first!")
                return

            
            transformed_text = vectorizer.transform([input_text])

            
            if hasattr(model, "predict_proba"):
                prediction_probs = model.predict_proba(transformed_text)[0]
            else:
                print("Model does not support probability predictions.")
                return

            
            sorted_indices = prediction_probs.argsort()[::-1]

            
            plt.figure(figsize=(10, 6))
            classes = [encoder.classes_[i] for i in sorted_indices]
            probs = [prediction_probs[i] * 100 for i in sorted_indices]

            bars = plt.bar(classes, probs, color='skyblue')
            plt.xlabel('Class')
            plt.ylabel('Probability (%)')
            plt.title('Prediction Probabilities')
            plt.xticks(rotation=45, ha='right')
            plt.tight_layout()

            for bar in bars:
                height = bar.get_height()
                plt.text(bar.get_x() + bar.get_width() / 2., height,
                         f'{height:.1f}%', ha='center', va='bottom')

            plt.show()

            
            print("Top 3 predictions:")
            for i in range(min(3, len(sorted_indices))):
                idx = sorted_indices[i]
                label = encoder.classes_[idx]
                probability = prediction_probs[idx] * 100
                print(f"{label}: {probability:.2f}%")

    analyze_button.on_click(on_button_click)

    
    display(text_input, analyze_button, output)
    print("Interactive tool is ready! Enter text and click 'Analyze'.")


create_rf_prediction_tool(rf_model, vectorizer, label_encoder)
```

```python
import joblib
import pandas as pd


rf_model = joblib.load("random_forest_model.pkl")  
nb_model = joblib.load("naive_bayes_model.pkl")    
vectorizer = joblib.load("tfidf_vectorizer.pkl")   
label_encoder = joblib.load("label_encoder.pkl")   


def predict_on_csv(input_csv_path, output_csv_path):
    
    df = pd.read_csv(input_csv_path)
    
    
    if "statement" not in df.columns:
        print("Error: The input CSV must have a 'statements' column.")
        return

   
    df.dropna(subset=["statement"], inplace=True)
    df = df[df['statement'].str.strip() != ""]

    
    if df.empty:
        print("Error: No valid statements found after cleaning.")
        return

    
    transformed_text = vectorizer.transform(df['statement'])

    
    rf_predictions = label_encoder.inverse_transform(rf_model.predict(transformed_text))
    nb_predictions = label_encoder.inverse_transform(nb_model.predict(transformed_text))

    
    df['Random Forest Prediction'] = rf_predictions
    df['Naive Bayes Prediction'] = nb_predictions

    
    df.to_csv(output_csv_path, index=False)
    print(f"Predictions saved to {output_csv_path}")


predict_on_csv("Combined_Data_Cleaned.csv", "predictions_output.csv")
```

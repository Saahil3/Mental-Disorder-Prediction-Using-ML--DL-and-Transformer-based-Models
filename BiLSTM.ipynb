```python
# Load preprocessed dataset
import pandas as pd
file_path = '/content/drive/MyDrive/CNN/reddit_mental_health_dataset_fengg1.csv'
df = pd.read_csv(file_path)
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.utils import resample
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score, f1_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# TensorFlow and Keras imports
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Dense, Embedding, LSTM, Bidirectional, Dropout, Input, Concatenate, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.optimizers import Adam

# Assuming df is your dataframe with 'clean_text' and 'disorder' columns
print("Initial Dataset Shape:", df.shape)
df.dropna(subset=['clean_text'], inplace=True)
print("After Dropping Missing Values:", df.shape)

# Encode disorders
encoder = LabelEncoder()
df['disorder_encoded'] = encoder.fit_transform(df['disorder'])
num_classes = len(encoder.classes_)
print(f"Number of classes: {num_classes}")
print(f"Classes: {list(encoder.classes_)}")

# Balance dataset
balanced_df = []
for label in df['disorder_encoded'].unique():
    samples = df[df['disorder_encoded'] == label]
    balanced_samples = resample(samples, replace=True, n_samples=4000, random_state=42)
    balanced_df.append(balanced_samples)
df_balanced = pd.concat(balanced_df)
print("After Balancing Dataset:", df_balanced.shape)

# Display class distribution
plt.figure(figsize=(12, 6))
sns.countplot(x='disorder', data=df_balanced)
plt.xticks(rotation=45, ha='right')
plt.title('Class Distribution After Balancing')
plt.tight_layout()
plt.show()

# Split data
X_train, X_test, y_train, y_test = train_test_split(
    df_balanced['clean_text'],
    df_balanced['disorder_encoded'],
    test_size=0.2,
    stratify=df_balanced['disorder_encoded'],  # Ensure stratified sampling
    random_state=42
)
print("Training Set Size:", X_train.shape)
print("Test Set Size:", X_test.shape)

# Convert labels to categorical for neural network
y_train_cat = to_categorical(y_train, num_classes=num_classes)
y_test_cat = to_categorical(y_test, num_classes=num_classes)

# Text preprocessing
MAX_NUM_WORDS = 30000  # Increased vocabulary size
MAX_SEQUENCE_LENGTH = 250  # Increased sequence length
EMBEDDING_DIM = 200  # Increased embedding dimension

# Create and fit tokenizer
tokenizer = Tokenizer(num_words=MAX_NUM_WORDS, oov_token='<OOV>')
tokenizer.fit_on_texts(X_train)
word_index = tokenizer.word_index
print(f"Found {len(word_index)} unique tokens.")

# Convert texts to sequences and pad
X_train_seq = tokenizer.texts_to_sequences(X_train)
X_test_seq = tokenizer.texts_to_sequences(X_test)

X_train_pad = pad_sequences(X_train_seq, maxlen=MAX_SEQUENCE_LENGTH, padding='post', truncating='post')
X_test_pad = pad_sequences(X_test_seq, maxlen=MAX_SEQUENCE_LENGTH, padding='post', truncating='post')

print(f"Training data shape after padding: {X_train_pad.shape}")
print(f"Testing data shape after padding: {X_test_pad.shape}")

# Calculate class weights for imbalanced data handling (optional)
from sklearn.utils.class_weight import compute_class_weight
class_weights = compute_class_weight('balanced', classes=np.unique(y_train), y=y_train)
class_weight_dict = {i: weight for i, weight in enumerate(class_weights)}

# Define a function to build a more advanced BiLSTM model
def build_advanced_bilstm_model():
    # Input layer
    inputs = Input(shape=(MAX_SEQUENCE_LENGTH,))

    # Embedding layer
    embedding = Embedding(input_dim=MAX_NUM_WORDS,
                         output_dim=EMBEDDING_DIM,
                         input_length=MAX_SEQUENCE_LENGTH,
                         mask_zero=True)(inputs)

    # BiLSTM layers with residual connections
    lstm1 = Bidirectional(LSTM(128, return_sequences=True))(embedding)
    lstm1 = BatchNormalization()(lstm1)
    lstm1 = Dropout(0.3)(lstm1)

    lstm2 = Bidirectional(LSTM(128, return_sequences=False))(lstm1)
    lstm2 = BatchNormalization()(lstm2)
    lstm2 = Dropout(0.3)(lstm2)

    # Dense layers with BatchNormalization
    dense1 = Dense(256, activation='relu')(lstm2)
    dense1 = BatchNormalization()(dense1)
    dense1 = Dropout(0.4)(dense1)

    dense2 = Dense(128, activation='relu')(dense1)
    dense2 = BatchNormalization()(dense2)
    dense2 = Dropout(0.3)(dense2)

    # Output layer
    outputs = Dense(num_classes, activation='softmax')(dense2)

    # Create model
    model = Model(inputs=inputs, outputs=outputs)

    # Compile model with Adam optimizer and learning rate
    optimizer = Adam(learning_rate=0.001)
    model.compile(loss='categorical_crossentropy',
                  optimizer=optimizer,
                  metrics=['accuracy', tf.keras.metrics.AUC()])

    return model

# Create the BiLSTM model
bilstm_model = build_advanced_bilstm_model()
print(bilstm_model.summary())

# Set up callbacks for better training
checkpoint = ModelCheckpoint(
    'best_bilstm_model.h5',
    monitor='val_loss',
    save_best_only=True,
    verbose=1
)

early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=5,
    restore_best_weights=True,
    verbose=1
)

reduce_lr = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=2,
    min_lr=0.00001,
    verbose=1
)

# Train the model with class weights
history = bilstm_model.fit(
    X_train_pad, y_train_cat,
    validation_split=0.15,
    epochs=20,
    batch_size=64,  # Larger batch size for faster training
    callbacks=[early_stopping, checkpoint, reduce_lr],
    class_weight=class_weight_dict,
    verbose=1
)

# Plot training history
def plot_training_history(history):
    # Plot training & validation accuracy
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'])
    plt.plot(history.history['val_accuracy'])
    plt.title('Model Accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Validation'], loc='upper left')

    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title('Model Loss')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Validation'], loc='upper left')

    plt.tight_layout()
    plt.show()

# Plot the training history
plot_training_history(history)

# Evaluate the model
y_pred_prob = bilstm_model.predict(X_test_pad)
y_pred = np.argmax(y_pred_prob, axis=1)
y_test_decoded = y_test.values

# Calculate metrics
accuracy = accuracy_score(y_test_decoded, y_pred)
f1 = f1_score(y_test_decoded, y_pred, average='weighted')

print("\nBiLSTM Model Performance:")
print(f"Accuracy: {accuracy:.4f}")
print(f"F1 Score (Weighted): {f1:.4f}")
print("\nClassification Report:")
print(classification_report(y_test_decoded, y_pred, target_names=encoder.classes_))

# Plot confusion matrix
def plot_confusion_matrix(y_true, y_pred, classes):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(14, 12))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=classes, yticklabels=classes)
    plt.title('Confusion Matrix')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

# Plot confusion matrix
plot_confusion_matrix(y_test_decoded, y_pred, encoder.classes_)

# Function to predict disorders with more detailed output
def predict_disorder(text, top_n=3):
    # Clean and preprocess the text (assuming you have a preprocessing function)
    # For this example, we'll use the raw text

    # Convert to sequence and pad
    sequence = tokenizer.texts_to_sequences([text])
    padded_sequence = pad_sequences(sequence, maxlen=MAX_SEQUENCE_LENGTH, padding='post', truncating='post')

    # Get prediction probabilities
    prediction = bilstm_model.predict(padded_sequence)[0]

    # Get top predictions
    top_indices = prediction.argsort()[-top_n:][::-1]
    results = []

    print(f"Top {top_n} predictions:")
    for idx in top_indices:
        disorder = encoder.classes_[idx]
        probability = prediction[idx] * 100
        results.append((disorder, probability))
        print(f"{disorder}: {probability:.2f}%")

    # Create a bar chart of predictions
    plt.figure(figsize=(10, 6))
    disorders = [result[0] for result in results]
    probs = [result[1] for result in results]

    sns.barplot(x=probs, y=disorders)
    plt.xlabel('Probability (%)')
    plt.ylabel('Disorder')
    plt.title('Disorder Prediction Probabilities')
    plt.xlim(0, 100)
    plt.tight_layout()
    plt.show()

    return results

# Save the model, tokenizer, and encoder for future use
import joblib

bilstm_model.save('bilstm_mental_disorder_model.h5')
joblib.dump(tokenizer, 'bilstm_tokenizer.joblib')
joblib.dump(encoder, 'disorder_label_encoder.joblib')
print("\nModel, tokenizer, and encoder saved successfully!")

# Example usage
print("\nTesting with sample text:")
sample_text = "I feel extremely anxious and worried all the time, I can't seem to control these feelings."
predict_disorder(sample_text)

# Adding another example with different symptoms
print("\nTesting with another sample:")
sample_text2 = "I haven't been able to sleep well for weeks. I feel so sad and empty inside. Nothing brings me joy anymore."
predict_disorder(sample_text2)

# Function to compare CNN and BiLSTM models (if you have both)
def compare_models(text, cnn_model, bilstm_model):
    # Preprocess for both models
    sequence = tokenizer.texts_to_sequences([text])
    padded_sequence = pad_sequences(sequence, maxlen=MAX_SEQUENCE_LENGTH, padding='post', truncating='post')

    # Get predictions
    cnn_pred = cnn_model.predict(padded_sequence)[0]
    bilstm_pred = bilstm_model.predict(padded_sequence)[0]

    # Get top disorder for each model
    cnn_top = encoder.classes_[np.argmax(cnn_pred)]
    bilstm_top = encoder.classes_[np.argmax(bilstm_pred)]

    # Print comparison
    print(f"Input text: {text}")
    print(f"CNN prediction: {cnn_top} ({np.max(cnn_pred)*100:.2f}%)")
    print(f"BiLSTM prediction: {bilstm_top} ({np.max(bilstm_pred)*100:.2f}%)")

    # Plot comparison
    plt.figure(figsize=(12, 6))

    # Get top 3 predictions for each model
    top_3_cnn = cnn_pred.argsort()[-3:][::-1]
    top_3_bilstm = bilstm_pred.argsort()[-3:][::-1]

    disorders_cnn = [encoder.classes_[i] for i in top_3_cnn]
    probs_cnn = [cnn_pred[i]*100 for i in top_3_cnn]

    disorders_bilstm = [encoder.classes_[i] for i in top_3_bilstm]
    probs_bilstm = [bilstm_pred[i]*100 for i in top_3_bilstm]

    plt.subplot(1, 2, 1)
    sns.barplot(x=probs_cnn, y=disorders_cnn)
    plt.title('CNN Model')
    plt.xlabel('Probability (%)')
    plt.xlim(0, 100)

    plt.subplot(1, 2, 2)
    sns.barplot(x=probs_bilstm, y=disorders_bilstm)
    plt.title('BiLSTM Model')
    plt.xlabel('Probability (%)')
    plt.xlim(0, 100)

    plt.tight_layout()
    plt.show()

# If you have both models loaded, you can uncomment this to compare them
# try:
#     cnn_model = tf.keras.models.load_model('cnn_mental_disorder_model.h5')
#     compare_models(sample_text, cnn_model, bilstm_model)
# except:
#     print("CNN model not found. Can't compare models.")
```
